```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading  the dataset
df = pd.read_csv('/content/Python-Corona (5) (2).csv')
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Basic information about the dataset
print(df.info())
```

```python
# Summary statistics of the dataset
print(df.describe())
```

```python
# Checking  for the  missing values
print(df.isnull().sum())
```

```python
# Visualize missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Handle missing values by filling with mean for numeric columns only
for column in df.select_dtypes(include=['number']):
    df[column].fillna(df[column].mean(), inplace=True)
```

```python
# Distribution of numerical features
df.hist(bins=30, figsize=(15, 10), color='blue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
# Identifying outliers using box plots
plt.figure(figsize=(15, 10))
for i, column in enumerate(df.select_dtypes(include=np.number).columns, 1):
    plt.subplot(4, 4, i)
    sns.boxplot(df[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```
